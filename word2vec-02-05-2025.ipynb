```python
# 1. 가장 유사한 단어 top 10 보기
model.wv.most_similar("love", topn=10)

# 2. 단어 연산 (king - man + woman ≈ queen)
model.wv.most_similar(positive=["king", "woman"], negative=["man"], topn=5)

# 3. 여러 단어 평균 유사 단어
model.wv.most_similar(["kiss", "hug", "heart"])

# 4. 단어 간 유사도 계산 (0~1 사이 float 값)
model.wv.similarity("love", "heart")

# 5. 제일 안 어울리는 단어 찾기
model.wv.doesnt_match(["love", "kiss", "heart", "gun"])

# 6. 단어 벡터 직접 확인
model.wv["love"]  # numpy array 출력
```
